```python
from keras.datasets import fashion_mnist
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

import pandas as pd
import numpy as np
```

```python
(trainX, trainy), (testX, testy) = fashion_mnist.load_data()
```

## Reduciendo el tamanio del dataset

```python
testX = testX[:1000, :].reshape(1000, 28*28)
testy = testy[:1000]
```

```python
testX.shape, testy.shape
```

```python
np.unique(testy)
```

## TSNE

```python
tsne = TSNE(n_components=2).fit_transform(testX)
```

```python
plt.scatter(tsne[:, 0], tsne[:, 1])
```

## Kmeans

```python
kmeans = KMeans(n_clusters=10, random_state=0).fit(tsne)
predictions = kmeans.labels_
```

```python
plt.scatter(tsne[:, 0], tsne[:, 1], c=predictions)
```

```python
plt.scatter(tsne[:, 0], tsne[:, 1], c=testy)
```

## Conclusiones

* Pareciera ser que la clusterizacion que se realizo sobre los datos reducidos con tse separo mejor los cluster que a primera vista.
* Los clusters formados con los labels verdaderos pareceiera que en la zona central se cruzan entre si, talvez la reduccion de dimensiones no logra separar de forma adecuada con solo 2 componentes.
